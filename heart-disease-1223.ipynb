```python
data=pd.read_csv("/kaggle/input/heart-disease/heart_disease.csv")
data.head(5)
```

```python
data.shape
```

```python
data.info()
```

```python
data.drop(columns=['Exercise Habits','Family Heart Disease','Sugar Consumption','Alcohol Consumption','Triglyceride Level','Fasting Blood Sugar','CRP Level'],inplace=True)
```

```python
data.isnull().sum()
```

```python
data['Age']=data['Age'].fillna(data['Age'].mean())
data['Blood Pressure']=data['Blood Pressure'].fillna(data['Blood Pressure'].mean())
data['Cholesterol Level']=data['Cholesterol Level'].fillna(data['Cholesterol Level'].mean())
data['BMI']=data['BMI'].fillna(data['BMI'].mean())
data['Sleep Hours']=data['Sleep Hours'].fillna(data['Sleep Hours'].mean())
data['Homocysteine Level']=data['Homocysteine Level'].fillna(data['Homocysteine Level'].mean())

```

```python
data['Gender']=data['Gender'].fillna(data['Gender'].mode()[0])
data['Smoking']=data['Smoking'].fillna(data['Smoking'].mode()[0])
data['Diabetes']=data['Diabetes'].fillna(data['Diabetes'].mode()[0])
data['High Blood Pressure']=data['High Blood Pressure'].fillna(data['High Blood Pressure'].mode()[0])
data['Low HDL Cholesterol']=data['Low HDL Cholesterol'].fillna(data['Low HDL Cholesterol'].mode()[0])
data['High LDL Cholesterol']=data['High LDL Cholesterol'].fillna(data['High LDL Cholesterol'].mode()[0])
data['Stress Level']=data['Stress Level'].fillna(data['Stress Level'].mode()[0])
```

```python
data.head(7)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
# one=data[['Gender','Smoking','Diabetes','High Blood Pressure','Low HDL Cholesterol','High LDL Cholesterol','Alcohol Consumption','Stress Level']]
```

```python
ohe=OneHotEncoder()
data['Gender']=ohe.fit_transform(data[['Gender']]).toarray()
data['Smoking']=ohe.fit_transform(data[['Smoking']]).toarray()
data['Diabetes']=ohe.fit_transform(data[['Diabetes']]).toarray()
data['High Blood Pressure']=ohe.fit_transform(data[['High Blood Pressure']]).toarray()
data['Low HDL Cholesterol']=ohe.fit_transform(data[['Low HDL Cholesterol']]).toarray()
data['High LDL Cholesterol']=ohe.fit_transform(data[['High LDL Cholesterol']]).toarray()
data['Stress Level']=ohe.fit_transform(data[['Stress Level']]).toarray()

```

```python
data.head(3)
```

```python
x=data.iloc[:,:-1]
y=data['Heart Disease Status']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
rf.score(x_test,y_test)*100
```
